```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[6,9,4,8,7 ]
plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="Step Plot")
plt.savefig("line1",dpi=2000,bbox_inches="tight",facecolor="y")
plt.title("Step Plot")
plt.xlabel("Maximum")
plt.ylabel("Minimum")
plt.legend()
plt.grid()
plt.show()
```
